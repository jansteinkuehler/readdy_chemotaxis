```python
import numpy as np
import readdy
import os
import sys
from datetime import datetime
from itertools import product
import matplotlib.pyplot as plt
from readdy.api.utils import load_trajectory_to_npy
```

```python
def plot_dist(file):
    t = readdy.Trajectory(file)
    if (os.path.getsize(file) > 300000000 ):
        s=100
    else:
        s=10

    n_particles_per_frame, positions, types, ids = load_trajectory_to_npy(
        file,begin=0,stride=s)
    
    plt.hist(positions[-1][0:15,1],bins=15)
    plt.figure()
    plt.plot(np.count_nonzero(positions[:,0:15,1]>0,axis=1)-7.5)
    plt.figure()
    plt.plot(n_particles_per_frame)
    plt.figure()
    plt.plot(np.mean(positions[:,0:15,1],axis=1))
    return np.mean(positions[:,0:15,1],axis=1)
```
